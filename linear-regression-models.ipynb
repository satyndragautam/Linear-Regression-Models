# **<p style="background-color:#0f5ca3;font-family:montserrat;color:#FFF9ED;font-size:140%;text-align:center;border-radius:15px 15px;">OVERVIEW</p>**

In this project, we will determine the **Linear Regression Models** and how it works to find the existence of an association relatishiop between a dependent (response variable or outcome variable or target variable) and independent variables (explanatory variable or predictor variable). Regression models do not establish a causal relationship between the dependent variable (Y) and the independent variable (X). In other words, using regression we cannot say that the value of Y depends on the value of X (or a change in the value of Y is caused due to a change in the value of X). We can only establish that the change in the value of Y is associated with the change in the value of X.

There are two types of linear regression model. 

* **Simple Linear Regression Model:** (a model with single independent variable, either one of them can be qualitative or quantitave variable)
* **Multiple Linear Regression Model** (a model with multiple independent variables)

**<p style = "color:#0f5ca3;text-align:center;font-style:italic">"The goal of these models is to create a linear model that minimizes the sum of squared errors/residuals(SSE)."</p>**

Before we dive into high-level details, we can simplyfy the model into key points;

* This model concerns two-dimensional sample points with one dependent and one independent variable. 
* The accuracy of each predicted value is measured by its squared residual. 
* The best fitting line passes through the centroid of the both variables Centroid: the point that is the intersection of the mean of each variables; X & Y. 
* A residual is a **vertical distance** between the points of data set (observed values) and fitted line values(predicted values).
* The goal is to make the squared error as small as possible. 
* Residual errors always add-up to zero. 

SLR implies that that there is only one independent variable in the model and the functional relationship between the dependent variable and the regression coefficient is linear. One of the functional forms of SLR is –

**Simple Linear Model Equation:** **Y**= b0+ b1**X** + e

where: <br>
b0 and b1 are known as the regression beta coefficients or parameters:
* b0 is the intercept of the regression line; that is the predicted value when X = 0. It can be simply calculated using the centroid values(average of both variables) and b1. 
* b1 is the slope of the regression line and it explains the average increase in Y associated with a one-unit increase in X.
* e is the error term (also known as the residual errors), the part of Y that cannot be explained by the regression model

we want to find an intercept b0 and a slope b1 such that the resulting line is as close as possible to the data points. There are a number of ways of measuring closeness. However, by far the most common approach involves minimizing the least squares criterion, and we take that approach in this project. In other words, the regression parameters are estimated using the method of **Ordinary Least Squares (OLS)**. In ordinary least squares, the objective is to find the optimal value of b0 and b1 that will minimise the Sum of Squares Error (SSE). The equation is used to predict a quantitative outcome Y on the basis of one single predictor variable X. Once, we build a statistically significant model, it is possible to use it for predicting future outcome on the basis of new X values. Based of the least squares method, we can calculated the b1 and b0 by using the following formula.



The variations casued in the model show how data points are spread from the actual mean of dependent variable or response variable and best fit line or predicted values. Below image explains how unexplained and explained variations are defined. 



* **Sum of Squares Total(SST):** It measures the total variation of Y values around the mean of Y. This is the numerator of the variation of Y.
* **Sum of Sqaures Error(SSE):** It measures the variation in Y attributable to factors other than relationship between X and Y or we can say it measures the unexplained variation.
* **Sum of Sqaures Regression(SSR):** It measures the amount of variability in the response(Y) that is explained (or removed) by performing the regression. 




**<p style = "color:#0f5ca3;text-align:center;font-style:italic">"The lower the SSE, the better the model."</p>**


**Co-efficient of Determination or R^2:** It measures the proportion of variability in Y that can be explained using X. An R^2 statistic that is close to 1 indicates that a large proportion of the variability in the response is explained by the regression. A number near 0 indicates that the regression does not explain much of the variability in the response; this might occur because the linear model is wrong, or the error variance σ^2 is high, or both. The R^2 statistic has an interpretational advantage over the SSE. Since unlike the SSE, it always lies between 0 and 1. However, it can still be challenging to determine what is a good R^2 value, and in general, this will depend on the application. In this case, we would expect to see if R2 value is close to 1 or not, and a substantially smaller R^2 value might indicate a serious problem with the model in which the data were generated.

### **<span style = "color:#0f5ca3">3.2 Validation of the SLR Model</span>**

It is important to validate the regression model to ensure its validity and goodness of fit before it can be used for practical applications. The following measures are used to validate the SLR model:

* **Coefficient of determination (R-square)**
* **Mean Square Error**
* **Standard Error**
* **Confidence Interval for slope and predivated values at the estimated value of X**
* **Confidence Interval for slop**
* **Confidence Interval for a given X value**
* **Prediction Interval for a given X value**
* **Hypothesis test for the regression coefficient (b1)**
* **Analysis of Variance for overall model validity (relevant for MLR)**
* **Residual analysis to validate the regression model assumptions**
* **Outlier analysis**

Now we will create the linear model using lm() function. 

```r
linear_model <- lm(mba_p~degree_p, data = mba_degree_percent)

linear_model

summary(linear_model)

#according to the linear regression model, we will determine the fitted values

pred_mba_value <- fitted.values(linear_model)
head(pred_mba_value)
```

* To check the accuracy of fitted model with actual sample data set, we will determine the mean value of both. We can also validate the model through residual calculation.

```r
mean(mba_degree_percent$mba_p) == mean(pred_mba_value) #The mean should be same

residual_of_model <- residuals(linear_model)
head(residual_of_model)

sum(residual_of_model) #must be zero
mean(residual_of_model) #must be zero
```

#### **Co-efficient of Determination (R-square):** 

From the above summary result, the R-square is **0.16** which means 16% of variability in MBA percentage is explained by the regression model and **84%** of the variabilty is unexplained due to other factors not included in the model. **Higher the unexplained variation, lower the fitness of model**. 

**NOTE:** A high R-square value is not necessarily a good indicator of the correctness of the model; it could be a spurious relationship. The adjusted R-squared is a statistic that modifies R2 by incorporating the sample size and number of explanatory variables.

#### **Mean Square Error:** 

It is an estimate of variance of the error(e) that shows how spread out the data points are from the regression line. It is SSE divided by its degree of freedom which is 2 in this example. 

**MSE** = **SSE**/**n-2**

where: <br>
SSE: sum of squared error

We will calculate the **SSE** now. 


```r
SSE <- sum((mba_degree_percent$mba_p - fitted(linear_model))^2)

print(SSE) #shows the value of SSE

SST <- sum((mean(mba_degree_percent$mba_p) - mba_degree_percent$mba_p)^2) #sum of squared total

print(SST)

SSR <- SST-SSE

print(SSR)

N <- mba_degree_percent %>%
    summarize(count = n())

print(N)

MSE <- SSE/(N-2)

print(MSE)
```

#### **Standard Error:** 
    It is the average distance a data point falls from the regression line in unit of the dependent variable. it measures how well the regression model makes predictions. It can be used to make the prediction intervals. It is also called ROOT MEAN SQUARE ERROR. 
    
**NOTE:** Standard error of the estimate is the standard deviation of the error term, (e). 

**<p style = "color:#0f5ca3;text-align:center;font-style:italic">"The lower standard error, the better the model."</p>**


```r

standard_error <- sqrt(MSE)

print(standard_error)
```

#### **Confidence Interval for slope and predivated values at the estimated value of X:** 
It shows the boundary of minimum and maximum value of intercept and slope at a certain value of confidence (alpha).

```r
confint(linear_model, level = 0.95)
```

#### **Confidence Interval for slope:**

We will determine the margin of error that will help us define the confidence interval for the slope;

Confidence interval for the slope b1(sβ1);



We will calculate the confidence interval for slope manually. We know that slope(b1) is for the model is **0.319**. Now we will have to calculate the margin of error at α = 0.025. The t-value at given significance level is **2.776**. The standard error of the estimate is **5.35**. 

```r
sum_squared_degree <- sqrt(sum((mba_degree_percent$degree_p - mean(mba_degree_percent$degree_p))^2))

stand_dev_slope <- 5.35/sum_squared_degree

margin_error <- stand_dev_slope * 2.774

confi_int_slope_upr <- 0.319 + margin_error

confi_int_slope_lwr <- 0.319 - margin_error

print(confi_int_slope_upr) #0.045
print(confi_int_slope_lwr) #0.18
```

#### **Hypothesis test for regression coefficient (t-Test):**

The regression coefficient b1 captures the existence of a linear relationship between the response variable and the explanatory variable. If b1 = 0, we can conclude that there is no statistically significant linear relationship between the two variables. <br>
The null and alternative hypothesis for the SLR model can be stated as follows:
* **Null:** There is no relationship between X and Y (b1 = 0)
* **Alternative:** There is a relationship between X and Y (b1 ≠ 0)

In our example, since the p-value for degree_p is less than 0.05, we reject the Null and conclude that there is significant evidence suggesting a linear relationship between degree_p (X) and mba_p (Y).

**OR** We can test our t-value(we will calculate) with the t-value at given α = 0.025 to determine if this regression is significant or not. 


```r
t_value_calculated <- 0.319/stand_dev_slope
print(t_value_calculated) #we can see the t-value for slope (6.41) is greater than t-value at α = 0.025 which is 2.78. Thus, the regression model is significant and there is relationship between the variables.
```

#### **Test for Overall Model: ANOVA (F-test):**

Using ANOVA, we can test whether the overall model is statistically significant. However, for a SLR, the null and alternative hypothesis in ANOVA and t-test are exactly same and thus there will be no difference in the p-value.
The Null and Alternative hypothesis for F-test are given by-<br>
* **Null:** There is no statistically significant relationship between Y and any of the explanatory variables (i.e all regression coefficients are zero)
* **Alternative:** Not all regression coefficients are zero

#### **Residual analysis:**

The difference between the observed value of the dependent variable (Y) and the predicted value (Y ̂, Y-hat) is called the residual (error). Each data point has one residual. Residual analysis is important to check whether the assumptions of regression models have been satisfied. It gives the best information about the error term, e. It is performed to check the following-
* The functional form of regression is correctly specified
* The expected value of error term is zero.
* For all the values of independent variable (X), the variance of the error term (e) is the same/constant (called homoscedasticity).
* The values of error term are independent to each other.
* The residuals (Y- Y ̂) follows a nomral distribution or bell-curve or bell-shaped distribution.
* If there are any outliers <br>
If residuals are evenly spread out, or uniformally distributed, it means that it is a good linear model. Regression diagnostic plots are used to check on the above.

### **<span style = "color:#0f5ca3">3.3 Diagnostic Plots</span>**

The diagnostic plots will define how accurate our hypothesis test is and how the evidence is supporting our linear model. 

**Note:** The four plots show the top 3 most extreme data points labelled with the row numbers of the data in the data set. You might want to take a close look at them individually to check if there is anything special for the observation. 

**_The standardized residual is the residual divided by its standard deviation._**

```r
#using R base package to create plots

plot(linear_model)
```

#### **Residuals vs Fitted:**
Used to check the linear relationship assumption. An approximately horizontal line (red line), without distinct patterns is an indication for a linear relationship. Any pattern in the residual plot would indicate incorrect specification of the model.<br>
In our example, there is no pattern in the residual plot. This suggests that we can assume linear relationship between the predictor (degree_p) and the outcome variable (mba_p).<br>

**Note:** If the residual plot indicates a non-linear relationship in the data, then a simple approach is to use non-linear transformations of the predictors, such as log(X), sqrt(X) and X2, in the regression model. 

#### **Normal Q-Q:** 
Used to examine whether the residuals are normally distributed. It is good if residuals points follow the straight dashed line.<br>
In our example, most of the points fall approximately along the reference line, so we can assume normality. Thus the hypothesis tests (t and F) are valid.<br>

**Note:** If the residual plot does not follow approximately normal distribution it puts a doubt on the validity of the model itself since we cannot trust the outcome of the hypothesis tests (p-value).

#### **Scale-Location(or)Spread-Location:**
Used to check the homogeneity of variance of the residuals (homoscedasticity). This plot shows if residuals are spread equally along the ranges of the predictors. Failure to meet this assumption will result in unreliability of the hypothesis tests. It is good if you see a horizontal line with equally spread points.   Horizontal line with equally spread points is a good indication of homoscedasticity.<br>

**Note:** If there is heteroscedasticity then we can expect a funnel type shape or other shapes in the residual plot. A funnel shape indicates that the variance of the residuals depends on the value of independent variable (X). The variance of residuals are not constant and error term is larger in the left side than in right side. If the residual analysis reveals heteroscedasticity, rebuild the model with different independent variable or fit a non-linear regression model but don't OVERFIT. We can use to justify the statistical tests such as **Breusch-Pagan Test, White Test, and Non-Constant Variance Test**

#### **Residuals vs Leverage:** 
Used to identify influential cases, that is, extreme values that might influence the regression results when included or excluded from the analysis. 

#### **Outliers:** 
Outliers are observations whose values show a large deviation from the mean value, that is (Y- Y ̅) is large. That is, an outlier is a data point whose response Y does not follow the general trend of the rest of the data. Presence of an outlier can have a significant influence on values of regression coefficients. Observations whose standardized residuals are greater than the ideal value of **3** are possible outliers. Formula we use to justify this statistical test is **Standardised residual = residual error / standard deviation of residual errors**. 


```r
outliners_check <- resi_main_data %>%
    mutate(stand_dev_resi_errors = sd(residual_of_model)) %>%
    mutate(standardized_resi = residual_of_model / stand_dev_resi_errors) %>%
    arrange(desc(standardized_resi)) %>%
    head(10)

print(outliners_check)
```

In oour case, as above result shows that we don't have any standardized residuals whose value is greater than **3**. It suggests that we don't have possible **outliners**. 


#### **High Leverage Points:** 
A data point that deviate a lot from the rest of the points for independent variable(X) is called **High Leverage Point**. These points literally work like a **lever** tilting the regression line towards that specific point. We can calculate the leverage value using the following formula; 




Leverage value of an observation measures the influence of that observation on the overall fit of the regression function. Leverage value of more than **decisive point = 3(p+1)/n** is treated as highly influential. We can calculate the decisive point using the following formula; 

**Decisive point = 3(p+1)/n** <br>
where: **p** = no. of independent variable <br>
       **n** = no. of data points

```r
#Now, first we will calculate the decisive point

decisive_point_df <- 3*(1+1)/N
print(decisive_point_df)

#We will compare which leverage points are greater than decisive point(0.028)

leverage_point_df <- resi_main_data %>%
    mutate(leverage_point = hatvalues(linear_model)) %>%
    mutate(decisive_point = c(0.028)) %>%
   filter(leverage_point >= decisive_point) %>%
    arrange(desc(leverage_point)) %>%
    head(10)

print(leverage_point_df)
```

# **<p style="background-color:#0f5ca3;font-family:montserrat;color:#FFF9ED;font-size:140%;text-align:center;border-radius:15px 15px;">MULTIPLE LINEAR REGRESSION MODEL</p>**

Multiple Linear Regression (MLR) is a statistical technique for finding existence of an association relationship between a dependent variable (response or outcome or target or criterion variable) and several independent variables (explanatory or predictor or regressor variable). It is an extension of **Simple Linear Regression**. It also allows us to determine the overall fit (variance explained) of the model and the relative contribution of each of the predictors to the total variance explained. The functional form of MLR is given by-

**Y = b0 + b1 * X1 + b2 * X2 + …. + bp * Xp + e**

where,
Y is the dependent variable, X1, X2 … Xp are independent variables, b0 is a constant, b1, b2 … bp are the partial regression coefficients corresponding to the explanatory variables and e is the error term (residual).<br>

In MLR, the regression co-efficients are called partial regression co-efficients since the relationship between an explanatory variable and the response variable is calculated after removing (partial out) the effect of all the other explanatory variables in the model. It is intepretated as the estimated change in Y corrosponding to one unit change in the independent variables when all other independent variables are held constant. 

* **
#### **Assumptions:**

Before using the multiple regression, the part of process involves **CHECKING** perscpective to make sure the data we need to analyze can actually be analyized or not. There are assumptions we need to make before using the model, 

* The dependent variable should be on a continuous scale or continuous data.
* The data must have two or more independent variables, that can be either continuous or categorical.
* There should be independence of observations or independence of residuals which can be checked by Durbin-Watson statistic tool. 
* There needs to be linear relationship between the dependent variable and independent variables.
* The data needs to show **homoscedasticity** which means the variance along the best fit line is spread out evenly. 
* The data must not show **multi-collinearity** between independent variables.
* There should be no significant **outliners** or **high leverage point/highly influential points** 
* The residual errors must be normally distributed.

**NOTE:** Adding more independent in your model doesn't mean that the regression will be better or offer better predictions. In fact, it can make things worse. This is normally called **Overfitting**. Also notice, adding more independent variables creates more relationships among them. There can be multi-collinearity between dependenet variable and independent variables but not between the independenet variables. 

* **


Before conduting the multiple regression, we also need to do a fair amount of prep-work; 

* Generate a list of potential variables
* Collect data on variables
* Check the relationship between each independent and dependent variable using scatterplots and correlations
* Check relationship between each independent variable
* Conduct simple linear regression for IV and DV pair 
* Use the non-reductant independent variables in the analysis to find out the best fitting model.

#### **VIF (Variance Inflation Factor):** 

It measures the severity of multi-collinearity in the multiple regression model analysis OR we can say it defines the multi-collinearity among the independent variabales. A large value of **VIF** indicates a highly collinear relationship among independent variables that should be adjusted for further prediction making. 

**It is equal to the ratio of the overall model variance to the variance of a model that includes only single independent variable in the model, and the ratio is calculated for each variable.** 

**Valuation:** <br>
**VIF = 1**, meaning **No Correlation** <br>
**VIF (1,5) or 1 <VIF< 5**, meaning **Moderate Correlation** <br>
**VIF > 5**, meaning **High Correlation** but as a thumb or rule, if it is 5 or 10, there would be problem. So 1-5 VIF is significant.

### **<span style = "color:#0f5ca3">4.2 Validation of the Multiple Regression Model</span>**

Before making a predictions based on our independent variables, we need to validate if the model is significant or not. All the evidence need to be gathered together to test our hypothesis test. 


<a id="MLR-two-1"></a>
#### **4.2.1 Validation of the overall regression model: F-Test**

The research question that we are answering here is, does the regression model contain at least one predictor variable useful in predicting the response variable.
Analysis of Variance (ANOVA) is used to validate the overall regression model. The null and alternative hypotheses are stated as- <br>

**Null: b0 = b1 = b2 = b3 = 0**<br>
**Alternative: Not all coefficients are zero**

In our example, it can be seen that p-value of the F-statistic is < **0.05**, which is highly significant and hence we reject the null hypothesis. This means that, at least one of the predictor variables is significantly related to the response variable.

The statement in alternative hypothesis is that not all the beta’s are zero, that is, some of the coefficients may be zero. This is the reason why we have to do the t-Test to check the existence of statistically significant relationship between the individual explanatory variables and the response variable.

**Note:** We usually don’t worry about the p-value for Constant. It has to do with the “intercept” of the model and seldom has any practical meaning unless it makes sense for all the independent variables to be zero simultaneously.

<a id="MLR-two-2"></a>
#### **4.2.2 Statistical significance of individual variables in MLR: t-Test**

Within a MLR model, we may want to know whether a particular independent variable is making a useful contribution to the model. That is, given the presence of the other independent variables in the model, does a particular variable help us predict or explain the dependent variable? <br>
For a given independent variable, the t-statistic evaluates whether or not there is significant association between the independent and the dependent variable, that is, whether the beta coefficient of the independent variable is significantly different from zero.<br>
The null and alternative hypotheses in the case of individual independent variables (Xi) and the dependent variable (Y) is stated as-
    
 **Null:** There is no relationship between independent variable (Xi) and dependent variable Y (bi = 0) <br>
 **Alternative:** There is a relationship between independent variable (Xi) and dependent variable Y (bi ≠ 0)
 
 
When we cannot reject the null hypothesis for a particular independent variable in the above, we should say that we do not need that variable (say X1) in the model given that variables X2 and X3 will remain in the model.<br>


To be more precise, we will now determine the t-test value (p-value) for each individual independent variable,

```r
mba_degree_model <- lm(mba_p~degree_p, data = train) #mba_p & degree_p

mba_ssc_model <- lm(mba_p~ssc_p, data = train) #for mba_p & ssc_p

mba_hsc_model <- lm(mba_p~hsc_p, data = train) #for mba_p & hsc_p

#so far p-value for independent variable and dependent variable is significant and it shows there is a relationship between individual variable and dependent variable

#Now we will check if there is collinear relationship between among the independent variables.

degree_ssc_model <- lm(degree_p~ssc_p, data = train)
summary(degree_ssc_model) #it shows high collinear relationship

degree_hsc_model <- lm(degree_p~hsc_p, data = train)
summary(degree_hsc_model) #it shows significantly less collinear relationship compared to other two independent variables.

ssc_hsc_model <- lm(ssc_p~hsc_p, data = train)
summary(ssc_hsc_model) #it shows high collinear relationship
```

#### **4.2.3 Coefficient of Multiple Determination (R-Square) and adjusted R-square**

**R-square** measures the proportion of variation in the dependent variable explained by the model. A problem with R-square is that it will always increase when more variables are added to the model even when there is no statistically significant relationship between the independent variable and the dependent variable. A solution to this is to adjust the value of R-square by taking into account the number of predictors. The Adjusted R-square reflects both the number of independent variables and the sample size and may increase or decrease when an independent variable is added or dropped, thus giving an indication of the value of adding or removing the independent variables in the model. 

An increase in Adjusted R-square indicates that the model has improved. The Adjusted R-square value will always be less than or equal to R-square value. <br>
In our example, the Adjusted R-square of 0.2328 means that 23.28% of the variability in MBA percentage is explained by the model. (This is better than the R-square of approximately 15.8% in the SLR model)<br>

#### **4.2.5 Multi-collinearity and Variance Inflation Factor**

When the data set has a large number of independent variables, it is possible that few of these independent variables may be highly correlated. Existence of high correlation between independent variables is called multi-collinearity. Presence of multi-collinearity can destabilise the MLR model. Due to presence of multi-collinearity it is possible that a statistically significant variable may be labelled as insignificant on account of inflated p-value or the sign of a regression coefficient may change. We can check **VIF** number to check if Multi-collinearity exists and how much it is. **As a rule of thumb, a VIF value that exceeds 5 or 10 indicates a problem.** VIF greater than this requires further investigation to assess the impact of multi-collinearity. Remedies to handle multi-collinearity include using **Principal Component Analysis (PCA), Ridge regression and LASSO regression**. <br>

### **<span style = "color:#0f5ca3">4.4 Cross-validation</span>**

How do we know that an estimated regression model is generalizable beyond the sample data used to fit it? 

One way is to partition the sample data into a training (or model-building) set, which we can use to develop the model, and a validation test (or prediction) set, which is used to evaluate the predictive ability of the model. This is called cross-validation. Cross-validation refers to a set of methods for measuring the performance of a given predictive model on new test data sets.<br>

The different cross-validation methods for assessing model performance are; <br>
* **The validation set approach**
* **Leave-one-out cross-validation**
* **k-fold cross-validation**
* **Repeated k-fold cross-validation**

Validation set approach involves randomly dividing the available set of observations into two parts, a training set and a validation data set (hold-out set). The model is fit on the training data set, and the fitted model is used to predict the responses for the observations in the validation test data set. The resulting validation set error rate – typically assessed using Root Mean Square Error (RMSE) provides an estimate of the test error rate.

* Sum of Square Errors (SSE) = ∑〖(Yi- Yi(hat))〗^2

* Mean Square Error (MSE) = SSE/n-3 (In this case)

* Root Mean Square Error (RMSE) = √MSE


**The lower the SSE, the better the model.**

In practice, the difference between RMSE and SSE is very small, particularly for large multivariate data.

If the predictions obtained using the model are very close to the actual values in the validation data set, then RMSE will be small, and we can conclude that the model fits the validation set very well. **The lower the RMSE, the better the model.**

Dividing the RMSE by the average value of the outcome variable will give us the prediction **error rate**, which should be **as small as possible**.

**Error Rate = RMSE(root mean square error) / Average of predicted values**

```r
SSE <- sum((test$mba_p - mba_p_pred)^2)

print(SSE) #shows the value of SSE

SST <- sum((mean(test$mba_p) - test$mba_p)^2) #sum of squared total

print(SST)

SSR <- SST-SSE

print(SSR)

N <- test %>%
    summarize(count = n())

print(N)

MSE <- SSE/(N-3)

print(MSE)

RMSE <- sqrt(MSE)

print(RMSE)

error_rate <- RMSE / mean(test$mba_p)

print(error_rate)
```
